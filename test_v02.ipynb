```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
```

```python
def trial_wave_function_1(x, a, B):

    if np.abs(x) <= a:
        return B * (np.square(a) - np.square(x))
    
    else:
        return 0
    
    # return B * (np.square(a) - np.square(x))



def trial_wave_function_2(x, beta, A):

    return A * np.exp(-1 * beta * np.square(x))


# def integrand_1(x, a, B):
#     if np.abs(x) <= a:
#         return np.square(trial_wave_function_1(x, a, B))
    
#     else:
#         return 0
    

# def integrand_2(x, beta, A):
#     return np.square(trial_wave_function_2(x, beta, A))


def probability_density_1(x, a, B, trial_wave_function_1):
    
    return np.square(trial_wave_function_1(x, a, B))



def probability_density_2(x, beta, A, trial_wave_function_2):
    
    return np.square(trial_wave_function_2(x, beta, A))
```

```python
a = 2
B = 1
beta = 2
A = 1


x_lim = 5

# intergation_1, _ = sp.integrate.quad(integrand_1, -1 * x_lim, 1 * x_lim, args=(a, B))
B = np.sqrt(15 / (16 * np.pow(a, 5))) #1 / np.sqrt(intergation_1)

# intergation_2, _ = sp.integrate.quad(integrand_2, -1 * np.inf, np.inf, args=(beta, A))
A = np.pow(((2 * beta) / np.pi), 0.25) #1 / np.sqrt(intergation_2)
# print(sp.integrate.quad(probability_density_2, -1 * np.inf, np.inf, args=(beta, A, trial_wave_function_2)))


# print(B)
# print(A)

x_arr = np.linspace(-1 * x_lim, 1 * x_lim, 1000001)
dx = x_arr[1] - x_arr[0]

probability_density_1_arr = np.zeros((len(x_arr)))
for i, x in enumerate(x_arr):
    probability_density_1_arr[i] = probability_density_1(x, a, B, trial_wave_function_1)

plt.plot(x_arr, probability_density_1_arr * dx, label=f'Sum: {(probability_density_1_arr * dx).sum()}')
# plt.ylim((0, 1))
plt.legend()
plt.grid()
plt.show()


probability_density_2_arr = np.zeros((len(x_arr)))
for i, x in enumerate(x_arr):
    probability_density_2_arr[i] = probability_density_2(x, beta, A, trial_wave_function_2)

plt.plot(x_arr, probability_density_2_arr * dx,  label=f'Sum: {(probability_density_1_arr * dx).sum()}')
# plt.ylim((0, 1))
plt.legend()
plt.grid()
plt.show()

# print((probability_density_1_arr * dx).sum())
# print((probability_density_2_arr * dx).sum())
```
