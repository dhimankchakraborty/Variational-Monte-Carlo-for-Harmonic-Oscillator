```python
def trial_wave_function_1(a, x):

    if np.abs(x) <= a:
        return (np.square(a) - np.square(x))
    
    else:
        return 0
    
    # return B * (np.square(a) - np.square(x))



def trial_wave_function_2(beta, x):

    return np.exp(-1 * beta * np.square(x))


def integrand_1(x, a):
    if np.abs(x) <= a:
        return np.square(trial_wave_function_1(a, x))
    
    else:
        return 0
    

def integrand_2(x, beta):
    return np.square(trial_wave_function_2(beta, x))


def probability_density_1(x, a, B, trial_wave_function_1):
    
    return np.square(trial_wave_function_1(a, x)) / B



def probability_density_2(x, beta, A, trial_wave_function_2):
    
    return np.square(trial_wave_function_2(beta, x)) / A
```

```python
a = 2
beta = 1.5

x_lim = 5

intergation_1, _ = sp.integrate.quad(integrand_1, -1 * x_lim, 1 * x_lim, args=(a))

intergation_2, _ = sp.integrate.quad(integrand_2, -1 * np.inf, np.inf, args=(beta))

print(intergation_1)
print(intergation_2)

x_arr = np.linspace(-1 * x_lim, 1 * x_lim, 1001)

probability_density_1_arr = np.zeros((len(x_arr)))
for i, x in enumerate(x_arr):
    probability_density_1_arr[i] = probability_density_1(x, a, intergation_1, trial_wave_function_1)

plt.plot(x_arr, probability_density_1_arr)
# plt.ylim((0, 1))
plt.grid()
plt.show()


probability_density_2_arr = np.zeros((len(x_arr)))
for i, x in enumerate(x_arr):
    probability_density_2_arr[i] = probability_density_2(x, beta, intergation_2, trial_wave_function_2)

plt.plot(x_arr, probability_density_2_arr)
# plt.ylim((0, 1))
plt.grid()
plt.show()
```
