```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from functions import *
```

```python
def local_energy_1(x, a):

    if np.abs(x) <= a:
        return (1 / (np.square(a) - np.square(x))) + np.square(x) / 2
    
    else:
        return 0



def local_energy_2(x, beta):
    return beta + (0.5 - 2 * np.square(beta)) * np.square(x)
```

```python
a = 2
beta = 2

x_arr = np.linspace(-5, 5, 101)
# print(x_arr)

local_energy_1_arr = np.zeros((len(x_arr)))
for i, x in enumerate(x_arr):
    local_energy_1_arr[i] = local_energy_1(x, a)

plt.plot(x_arr, local_energy_1_arr)
plt.grid()
plt.show()


local_energy_2_arr = np.zeros((len(x_arr)))
for i, x in enumerate(x_arr):
    local_energy_2_arr[i] = local_energy_2(x, beta)

plt.plot(x_arr, local_energy_2_arr)
plt.grid()
plt.show()
```
